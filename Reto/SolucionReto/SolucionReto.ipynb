##1. Introduction: Explanation and Understanding of the Problem to Address

The "Titanic: Machine Learning from Disaster" challenge on Kaggle aims to predict the survival of passengers aboard the RMS Titanic, based on a set of provided features.   - Model accuracy: **0.7942583732057417%**

  - Model accuracy: **0.784688995215311%**
  
- **Merge**, combines titles and KNN.
  - Model accuracy: **0.7942583732057417%**
  
- **Median** (median of all values).
  - Model accuracy: **0.7990430622009569%**

Due to the better precision provided, we chose the **"Median"** method for imputing **Age**.

##### The models we will test are:
- Decision Trees: A simple, interpretable model that splits the data based on feature values to make decisions. Useful to identify some patterns in the data.
- Random Forest: An ensemble of decision trees that reduces overfitting by averaging the predictions of multiple trees. It‚Äôs more robust and generally performs better than a single decision tree.
- Logistic Regression: A simple linear model that predicts the probability of an outcome. It‚Äôs a good baseline model for binary classification problems.

**Imputation based on title, class and gender**

```python
def impute_age_title(data):

    # Make sure that the column 'Name' exists in the DataFrame.
    if 'Name' not in data.columns:
        raise KeyError("The column 'Name' does not exist in the DataFrame.")

    # Convert all strings in the 'Name' column to lowercase
    data['Name'] = data['Name'].str.lower()

    # Function to extract the title from the name
    def give_title(row):
        title = str(re.split(r'[,.]+', row['Name'])[1]).strip()
        return title

    # Applying the give_title function to extract titles
    data['Title'] = data.apply(give_title, axis=1)

    # Replace equivalent titles to unify
    data.replace({'Title': {'mlle': 'miss', 'ms': 'miss', 'mme': 'mrs'}}, inplace=True)

    label_encoder = LabelEncoder()
    data['Title'] = label_encoder.fit_transform(data['Title'])

    # Function to impute age based on title, class and gender
    def give_age(row, df):
        age = row['Age']
        if pd.isna(age):
            title = row['Title']
            pclass = row['Pclass']
            sex = row['Sex']
            # Impute using the median of 'age' based on 'title', 'pclass', and 'sex'.
            age = df[(df['Title'] == title) & (df['Pclass'] == pclass) & (df['Sex'] == sex)]['Age'].median()
            # If no median is found, try with only 'pclass' and 'sex'.
            if pd.isna(age):
                age = df[(df['Pclass'] == pclass) & (df['Sex'] == sex)]['Age'].median()
        return age

    # Apply the give_age function to each row of the DataFrame
    data['Age'] = data.apply(lambda row: give_age(row, data), axis=1)

    return data
```

```python
# New preprocess data with new imputation function
def age_imp_preprocess_data_title(data):
    data = impute_age_title(data)
    data.drop(['Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
    data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
    data['Fare'].fillna(data['Fare'].median(), inplace=True)
    data['Sex'] = data['Sex'].map({'female': 1, 'male': 0})
    data = pd.get_dummies(data, columns=['Embarked'])
    return data
```

## **Feature engineering**

The model is capturing the 'Sex' interactions very well, therefore we need to add some variables to uncommon interactions to assume.

Variables added and their porpuse:
- Family: capture 'SibSp' and 'Parch' in one variable.
- Alone+Pclass1: people traveling alone in first class. Why? because people traveling alone probably are more likely to die. But what could be the exception? people who were in 1st class.
- Alone+Pclass3: Capture the group that are more likely to die.

Variables deleted:
- Deleted 'SibSp' and 'Parch' to avoid multicolinearity with 'Family'.

```python
def fe_preprocess_data(data):
    data['Age'].fillna(data['Age'].median(), inplace=True)
    data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
    data['Sex'] = data['Sex'].map({'female': 1, 'male': 0})
    data["Fare"].fillna(data.groupby("Pclass")["Fare"].transform("median"), inplace=True)

    data["Age"] = data["Age"].apply(lambda x: x - 30)

    data["Family"] = np.where(data["SibSp"] + data["Parch"] > 0, "With Family", "Alone")
    data["Alone+Pclass1"] = np.where((data["Family"] == "Alone") & (data["Pclass"] == 1), 1, 0)
    data["Alone+Pclass3"] = np.where((data["Family"] == "Alone") & (data["Pclass"] == 3), 1, 0)

    data.drop(['Name', 'Ticket', 'Cabin', "SibSp", "Parch"], axis=1, inplace=True)

    data = pd.get_dummies(data, columns=['Embarked'])

    return data
```

## **Hyperparameter Tuning**

We are going to find the best parameters for the model using Optuna.

Optuna does not use traditional grid search. Instead, it uses a more advanced and efficient approach called sequential model-based optimization (SMBO), with a focus on Tree-structured Parzen Estimator (TPE) as its default sampler.

Optuna uses TPE to model the objective function. Instead of randomly selecting hyperparameters, it builds a probabilistic model that represents the relationship between hyperparameters and performance metrics (e.g., accuracy or loss).

By using this method, Optuna avoids the exhaustive and time-consuming nature of grid search and is more effective than random search.



```python
def objective(trial):
    # Define the hyperparameter search space
    params = {
        'depth': trial.suggest_int('depth', 1, 16),
        'learning_rate': trial.suggest_float('learning_rate', 1e-4, 0.5, log=True),
        'iterations': trial.suggest_int('iterations', 100, 1000),
        'l2_leaf_reg': trial.suggest_float('l2_leaf_reg', 1e-2, 10, log=True),
        'border_count': trial.suggest_int('border_count', 1, 255),
        'random_strength': trial.suggest_float('random_strength', 1e-9, 10, log=True),
        'rsm': trial.suggest_float('rsm', 0.1, 1.0),
        'scale_pos_weight': trial.suggest_float('scale_pos_weight', 0.1, 10),
        'boosting_type': trial.suggest_categorical('boosting_type', ['Ordered', 'Plain']),
        'bootstrap_type': trial.suggest_categorical('bootstrap_type', ['Bayesian', 'Bernoulli', 'MVS', 'No']),
    }

    # Conditionally set bootstrap-related parameters
    if params['bootstrap_type'] == 'Bayesian':
        params['bagging_temperature'] = trial.suggest_float('bagging_temperature', 0.0, 10.0)
    elif params['bootstrap_type'] in ['Bernoulli', 'MVS']:
        params['subsample'] = trial.suggest_float('subsample', 0.1, 1.0)

    # Initialize the model with the suggested hyperparameters
    model = CatBoostClassifier(
        loss_function='Logloss',
        eval_metric='Accuracy',
        verbose=False,
        random_seed=42,
        **params
    )

    # Fit the model
    model.fit(Pool(X_train, y_train, cat_features=cat_features), eval_set=(X_test, y_test))

    # Predict and calculate accuracy
    preds = model.predict(X_test)
    accuracy = accuracy_score(y_test, preds)

    return accuracy

# Optimize using Optuna
study = optuna.create_study(direction='maximize', sampler=optuna.samplers.TPESampler(seed=10))
study.optimize(objective, n_trials=100)

# Extract the best hyperparameters
best_params = study.best_params

# Train the final model with the best hyperparameters
best_model = CatBoostClassifier(
    loss_function='Logloss',
    eval_metric='Accuracy',
    random_seed=42,
    verbose=False,
    **best_params
)

best_model.fit(X_train, y_train, cat_features=cat_features, eval_set=(X_test, y_test))
best_y_pred = best_model.predict(X_test)

# Evaluate the final model
final_accuracy = accuracy_score(y_test, best_y_pred)
print(f"Best accuracy: {final_accuracy}")
```
